```python
from tweepy import API 
from tweepy import OAuthHandler
```

```python

ACCESS_TOKEN = '####################################'
ACCESS_TOKEN_SECRET = '####################################'
CONSUMER_KEY = '####################################'
CONSUMER_SECRET = '####################################'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# # # TWITTER CLIENT # # #
class TwitterClient():
    
    def __init__(self, twitter_user=None):
        self.auth = TwitterAuthenticator().authenticate_twitter_app()
        self.twitter_client = API(self.auth)

        self.twitter_user = twitter_user

    def get_twitter_client_api(self):
        return self.twitter_client
```

```python
# # # # TWITTER AUTHENTICATER # # # #
class TwitterAuthenticator():

    def authenticate_twitter_app(self):
        auth = OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
        auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
        return auth
```

```python
twitter_client = TwitterClient()

api = twitter_client.get_twitter_client_api()

tweets = api.search(q="#", lang="en",count=60 , tweet_mode='extended')


df = pd.DataFrame()

Tweets = []
Tweet_length = []
Is_RT = []
Id = []
Date = []
Source = []
Retweets = []
for i in range(40):
    try:
        x = tweets[i].retweeted_status.full_text
        Is_RT.append(1)
    except:
        x = tweets[i].full_text
        Is_RT.append(0)
    Tweets.append(x)
    Tweet_length.append(len(x))
    Id.append(tweets[i].id)
    Date.append(tweets[i].created_at)
    Source.append(tweets[i].source)
    Retweets.append(tweets[i].retweet_count)
```

```python
df['Tweet'] = Tweets
df['Length'] = Tweet_length
df['Is_RT'] = Is_RT
df['ID'] = Id
df['Date'] = Date
df['Source'] = Source
df['Retweets'] = Retweets
```
